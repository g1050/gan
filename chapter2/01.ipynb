# GAN 生成对抗网络
2014年Goodfellow提出,20年来机器学习最酷的想法
生成模型（VAE,GAN）、判别模型
GAN：生成器学习真实数据的分布、判别器,二者对抗，最后达到一个平衡状态，最后判别器的正确率0.5，相当于随机判别
# 算法流程
G:输入一个随机噪声z，通过噪声生成图片,生成器的损失是通过判别器来计算的
D，输入x，输出它是不是真实的图片，判别器的损失函数，用交叉熵计算
公式：  
\[ L_D = -\mathbb{E}_{x \sim p_{\text{data}}(x)}[\log D(x)] - \mathbb{E}_{z \sim p(z)}[\log(1 - D(G(z)))] \]  
  
D(x)，放大损失，D(x)接近于0的时候，logD(x)是非常小的值，接近1的时候logd(x)接近0  
前半部分，用判别器判别，希望判别器判定为真，希望d(x)判别为1
